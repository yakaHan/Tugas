```
import csv
import numpy as np
import matplotlib.pyplot as plt
import statistics as st

with open("tugas1.csv", newline='') as csvfile:
    csv_reader = csv.DictReader(csvfile)

data5022 = []
with open("tugas1.csv", newline='') as csvfile:
    csv_reader = csv.reader(csvfile)
    for row in csv_reader:
        data5022.append(int(row[0]))

mean = np.mean(data5022)
mean_deviation = np.mean(np.abs(data5022 - mean))
print(mean_deviation)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of Data')
plt.grid(True)
plt.plot(data5022)
plt.show()
```
