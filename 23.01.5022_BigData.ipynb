##Tugas 1: Analisis Dataset Iris
Gunakan dataset iris dari Seaborn untuk melakukan EDA:
1. Hitung mean, median, modus, dan standar deviasi untuk setiap fitur (sepal_length,
sepal_width, petal_length, petal_width).
2. Buat scatter plot untuk memvisualisasikan hubungan antara sepal_length dan
petal_length.
3. Buat heatmap untuk menunjukkan korelasi antar fitur.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

iris = sns.load_dataset('iris')
print(iris.head())
```

```python
# Menghitung mean, median, modus dan standar deviasi
# Mean
mean_vals = iris.mean(numeric_only=True)
print("Mean:\n", mean_vals)

# Median
median_vals = iris.median(numeric_only=True)
print("\nMedian:\n", median_vals)

# Modus
mode_vals = iris.mode(numeric_only=True).iloc[0]
print("\nModus:\n", mode_vals)

# Standar Deviasi
std_vals = iris.std(numeric_only=True)
print("\nStandar Deviasi:\n", std_vals)
```

```python
# Scatter plot
plt.figure(figsize=(8,6))
sns.scatterplot(data=iris, x='sepal_length', y='petal_length', hue='species')
plt.title('Scatter Plot: Sepal Length vs Petal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.grid(True)
plt.show()
```

```python
# Korelasi antar fitur
correlation = iris.corr(numeric_only=True)

# Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap Korelasi')
plt.show()
```

Cari dataset publik (misalnya dari Kaggle) dan lakukan EDA lengkap:

2. Buat minimal 3 visualisasi yang relevan.

```python
df = pd.read_csv("/content/drive/MyDrive/disney_plus_titles.csv")
```

```python
print("Informasi Dataset:")
print(df.info())
print("\nStatistik Deskriptif:")
print(df.describe(include='all'))
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='type')
plt.xticks(rotation=90)
plt.xlabel('type')
plt.ylabel('Jumlah Film')
plt.title('Perbandingan Jumlah Type Movie dan TV Show Disney+')
plt.show()
```

```python
plt.figure(figsize=(12,5))
df['release_year'].value_counts().sort_index().plot(kind='bar')
plt.title('Tahun Rilis Film dan Series')
plt.xlabel('rating')
plt.ylabel('Tahun')
plt.tight_layout()
plt.show()
```

```python
top_countries = df['rating'].value_counts().head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_countries.values, y=top_countries.index)
plt.title('10 Rating terbanyak di tonton')
plt.xlabel('Jumlah Tayangan')
plt.ylabel('Rating')
plt.show()
```
